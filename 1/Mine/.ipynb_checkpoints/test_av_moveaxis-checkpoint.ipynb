```python
import numpy as np
```

```python
y = np.zeros((2,3,4))
print(y)

print("-------")
print(y.shape)
print("moveaxis(y,0,-2)")
print("------------")
z = np.moveaxis(y,0,-2)
print(z)
print(z.shape)
print("----------")
```

```python
j = np.ones((2,3,4,2))
print(j)
print(j.shape)
k = np.moveaxis(j, 1, -2)
# tippe k.shape blir 2,4,3,2
print("------")
print(k)
print(k.shape)
```

```python
a = np.ones((2,3,4,5))
print(a)
print(a.shape)
b = np.moveaxis(a, 1, 3)
# b.shape blir 2,4,5,3
print("------")
print(b)
print(b.shape)
```

```python
t = np.ones((2,3,4,5,6,7,8,9))
u = np.moveaxis(t, 2, 5)
print("t.shape", t.shape)
print("u.shape",u.shape)
```
